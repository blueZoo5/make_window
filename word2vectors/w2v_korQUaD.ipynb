```python
# 태그 제거 함수
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
try:
    # 텍스트 파일 읽기
    with open(txt_path, 'r', encoding='utf-8') as txt_file:
        text_content = txt_file.read()

    # HTML 태그 제거 함수 적용
    clean_text = remove_html_tags(text_content)

    # HTML 태그가 제거된 텍스트 파일로 저장하기
    with open(txt_path, 'w', encoding='utf-8') as txt_file:
        txt_file.write(clean_text)

    print(f"HTML 태그가 제거된 텍스트가 '{txt_path}' 파일로 저장되었습니다.")
except FileNotFoundError:
    print(f"에러: 텍스트 파일 '{txt_path}'을(를) 찾을 수 없습니다.")
except Exception as e:
    print(f"알 수 없는 에러가 발생했습니다: {e}")
```

```python
from tqdm import tqdm
from konlpy.tag import Mecab 
```

```python
mecab = Mecab('C:\mecab\mecab-ko-dic')
```

```python
len(clean_text)
```

```python
result = []

for line in tqdm(range(100)):
  # 소요시간 문제로 일부만 수행
  if line:
    result.append(mecab.morphs(clean_text))
```

```python
from gensim.models import Word2Vec
model = Word2Vec(result, window=5, min_count=5, workers=4, sg=0)
```

```python
# 적은 학습데이터로 원하는 모델 생성 실패

model_result1 = model.wv.most_similar("한국")
print(model_result1)
```
