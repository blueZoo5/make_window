```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/USER/Desktop/project/ML_study/archive/CommentsMarch2017.csv')
df.head(5)
```

```python
cmmt_df = pd.DataFrame({'commentBody': df['commentBody']})
# 특수 문자 제거
cmmt_df['clean_doc'] = cmmt_df['commentBody'].str.replace("[^a-zA-Z]", " ")
# 길이가 3이하인 단어는 제거 (길이가 짧은 단어 제거)
cmmt_df['clean_doc'] = cmmt_df['clean_doc'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
# 전체 단어에 대한 소문자 변환
cmmt_df['clean_doc'] = cmmt_df['clean_doc'].apply(lambda x: x.lower())
```

```python
cmmt_df
```

```python
#단어 토크나이징
tokenized_cmmt = cmmt_df['clean_doc'].apply(lambda x: x.split())
tokenized_cmmt = tokenized_cmmt.to_list()
```

```python
tokenized_cmmt
```

```python
from gensim.models.fasttext import FastText
import gensim.models.word2vec
```

```python
# Festtext 모델 생성
model = FastText(vector_size=100, window=5, min_count=1)  # 모델 초기화
model.build_vocab(tokenized_cmmt)  # 단어 사전 구축
model.train(tokenized_cmmt, total_examples=len(tokenized_cmmt), epochs=6)  # 모델 학습
```

```python
# 모델 저장
model.save("fasttext_cmmt2017.model")

# 모델 로드
model = FastText.load("fasttext_cmmt2017.model")
```

```python
word_vector = model.wv['review']
print("Word Vector for 'review':", word_vector)
```

```python
# 유사한 단어 찾기
similar_words = model.wv.most_similar('review', topn=5)
print("Most similar words to 'review':", similar_words)
```

```python
similar_words = model.wv.most_similar('news', topn=5)
print("Most similar words to 'review':", similar_words)
```
