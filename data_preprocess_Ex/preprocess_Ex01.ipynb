```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_excel('data/Data05.xlsx', skiprows = 1)
df1
```

```python
df1.head(10)
```

```python
print(df1.shape)
print(df1.columns)
```

```python
# 제품코드와 제품명이 키가 될 것을 고려하여 재고일자와 재고수량을 각각 하나의 칼럼으로 지정

df2 = pd.DataFrame(df1.iloc[:, 5:-2].drop(columns = ['Unnamed: 7', '판매']).set_index(['제품코드', '제품명']).stack()).reset_index()
df2.rename(columns = {'level_2' : '재고일자', 0 : '재고수량'})
```

```python
# 추가적으로 필요한 칼럼들을 가져오기(카테고리명, 재자그룹명, 제품코드, 제품명, 안전재고, 분류)

df3 = df1[['카테고리명', '자재그룹명', '제품코드', '제품명', '안전재고', ' 분류']]
df3
```

```python
# df2와 df3 병합
df4 = pd.merge(df3, df2, on = ['제품코드', '제품명'], how = 'left')
df5 = df4.rename(columns = {'level_2' : '재고일자', 0 : '재고수량'})
```

```python
df5
```

```python
df5.to_excel('preprocessed_data05.xlsx')
```

```python
df5.info()
```

```python
#제품별 재고수량 확인

pd.pivot_table(data=df5,index='제품명', values='재고수량',aggfunc='sum')
```

```python
#제품별 재고수량 확인

plt.figure(figsize = [10, 5])

sns.barplot(data = df5, x = '제품명', y = '재고수량')
```

```python
#품목이 너무 많은 관계로 재고수량 상위 10개 품목 확인

p1 = pd.pivot_table(data = df5, index = '제품명', values = '재고수량', aggfunc = 'sum')
top_ls = p1.sort_values(by = '재고수량', ascending = False).head(10).index.tolist()

top_ls
```

```python
# 상위 10개 품목의 재고수량 그래프

plt.figure(figsize = [10, 5])
plt.xticks(rotation=30, ha='right')
sns.barplot(data = df5, x = '제품명', y = '재고수량', order = top_ls, ci = None, estimator = sum)
```

```python
# 연도 및 월별 재고량 파악을 위한 작업

df5['year'] = df5['재고일자'].dt.year
df5['month'] = df5['재고일자'].dt.month
df5.head(10)
```

```python
#월별 전체 제품의 재고수량 확인

plt.figure(figsize = [10, 5])
sns.barplot(data = df5, x = 'month', y = '재고수량', ci = None, estimator = sum)
```

```python
#연도별 전체 제품의 재고수량 확인

plt.figure(figsize = [10, 5])
sns.barplot(data = df5, x = 'year', y = '재고수량', ci = None, estimator = sum)
```
